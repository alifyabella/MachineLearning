# Latihan 1

1. Bagaimana persepsi Marry saat temperatur udara 16 derajat celsius dan kecepatan 3km/jam?

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from math import sqrt
```

```python
#dataset
data = {
    'Temperatur': [10,25,15,20,18,20,22,24],
    'Kecepatan': [0,0,5,3,7,10,5,6],
    'Label': ['Dingin','Panas','Dingin','Panas','Dingin','Dingin','Panas','Panas']
}

df = pd.DataFrame(data)
df
```

Latih model KNN

```python
#encode label (Panas = 1, Dingin = 0)
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Label'])

X = df[['Temperatur', 'Kecepatan']]
y = df['Label']
```

2. Tentukan nilai k yang tepat berdasarkan akurasi yang paling baik

```python
akurasi_k = {}

for k in range(1, 9):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X, y)
    pred = model.predict(X)
    akurasi = accuracy_score(y, pred)
    akurasi_k[k] = akurasi

akurasi_k
```

ambil k terbaik

```python
k_terbaik = max(akurasi_k, key=akurasi_k.get)
k_terbaik
```

Prediksi data uji: Temperatur 16 derajat celcius dan kecepatan 3 km/jam

```python
model = KNeighborsClassifier(n_neighbors=k_terbaik)
model.fit(X, y)

test = [[16, 3]]

prediksi = model.predict(test)
hasil = le.inverse_transform(prediksi)[0]

print("Persepsi Marry:", hasil)
print("k terbaik =", k_terbaik)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

data2 = {
    'NIM': ['TI001','TI002','TI003','TI004','TI005','TI006','TI007','TI008','TI009','TI010'],
    'Actual': ['Lulus','Lulus','Lulus','Lulus','Lulus','Tidak Lulus','Tidak Lulus','Tidak Lulus','Tidak Lulus','Tidak Lulus'],
    'Prediksi': ['Lulus','Tidak Lulus','Lulus','Tidak Lulus','Tidak Lulus','Lulus','Tidak Lulus','Tidak Lulus','Tidak Lulus', 'Tidak Lulus']
}

df2 = pd.DataFrame(data2)
df2
```

```python
data = {
    'Temperatur': [14.0, 39.0, 30.0, 38.0, 27.0, 32.0, -2.0, 3.0, 3.0, 28.0],
    'Humidity': [73, 96, 64, 83, 74, 55, 97, 85, 83, 74],
    'Wind Speed': [9.5, 8.5, 7.0, 8.3, 17.0, 3.5, 8.0, 6.0, 6.0, 8.5],
    'Precipitation (%)': [82.0, 71.0, 16.0, 82.0, 66.0, 26.0, 86.0, 96.0, 66.0, 107.0],
    'Cloud Cover': ['partly cloudy','partly cloudy','clear','clear','overcast','overcast','overcast','partly cloudy','overcast','clear'],
    'Atmospheric Pressure': [1010.82, 1011.43, 1018.72, 1026.25, 990.67, 1010.03, 990.87, 984.46, 999.44, 1012.13],
    'UV Index': [2, 7, 5, 7, 1, 2, 1, 1, 0, 8],
    'Season': ['Winter','Spring','Spring','Spring','Winter','Summer','Winter','Winter','Winter','Winter'],
    'Visibility (km)': [3.5, 10.0, 5.5, 1.0, 2.5, 5.0, 4.0, 3.5, 1.0, 7.5],
    'Location': ['inland','inland','mountain','coastal','mountain','inland','inland','inland','mountain','coastal'],
    'Weather Type': ['Rainy','Cloudy','Sunny','Sunny','Rainy','Cloudy','Snowy','Snowy','Snowy','Sunny']
}

df3 = pd.DataFrame(data)

print("=== Dataset Cuaca ===")
print(df.head(), "\n")
```

Preprocessing

```python
#pisahkan fitur dan label
X = df3.drop('Weather Type', axis=1)
y = df3['Weather Type']

#encode fitur kategori
label_encoders = {}
for col in X.select_dtypes(include=['object']).columns:
  le = LabelEncoder()
  X[col] = le.fit_transform(X[col])
  label_encoders[col] = le

#encode label output
label_y = LabelEncoder()
y = label_y.fit_transform(y)

#normalisasi
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

Split Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Training KNN

```python
k = 3
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_train, y_train)
```

Evaluasi

```python
y_pred = model.predict(X_test)

print("=== Akurasi Model ===")
print("Akurasi:", round(accuracy_score(y_test, y_pred), 3), "\n")

print("=== Classification Report ===")

# Ambil kelas yang muncul pada data uji
classes_in_test = np.unique(y_test)

# Ambil nama kelas sesuai encoder
target_names = label_y.inverse_transform(classes_in_test)

print(classification_report(
    y_test,
    y_pred,
    labels=classes_in_test,
    target_names=target_names,
    zero_division=0
))
```

Contoh prediksi baru

```python
sample = np.array([[20, 80, 6.0, 50, 'clear', 1012, 5, 'Winter', 5.0, 'inland', 'Unknown']], dtype=object)
sample_df3 = pd.DataFrame(sample, columns=df3.columns)
```

Encode fitur kategori

```python
for col_name in label_encoders.keys():
  if col_name in sample_df3.columns:
    le = label_encoders[col_name]
    sample_df3[col_name] = le.transform(sample_df3[col_name])
```

Scaling

```python
sample_features = sample_df3.drop('Weather Type', axis=1)
sample_scaled = scaler.transform(sample_features)

hasil = model.predict(sample_scaled)
prediksi_cuaca = label_y.inverse_transform(hasil)

print("\n=== Prediksi Cuaca Baru ===")
print("Hasil prediksi:", prediksi_cuaca[0])
```
